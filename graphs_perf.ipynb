```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./training/logs_2022-07-02_23h57")
df2 = pd.read_csv("./training/logs_2022-07-03_13h21")
```

```python
df2["episode"] += 500
```

```python
frames = [df, df2]
df = pd.concat(frames)
```

```python
df2["episode"]
```

```python
df.tail()
```

```python
dff = df[['episode', 'test_success_ratio', 'epsilon', 'test_score', 'train_score']]
```

```python
sns.lmplot(x='episode', y='test_score', data=dff,
           order=2, ci=None, scatter_kws={"s": 80}).set(title="Average reward for every 10 episodes")
```

```python
sns.lmplot(x='episode', y='test_success_ratio', data=df2,
           order=1, ci=None, scatter_kws={"s": 40}).set(title="Average steps for every 10 episodes ")
```
